```python
#Upload the file to google colab
#Open the file for reading and create a file object
#Handle FileNotFoundException
lines_list =[]
try:
  fh = open("gasprices.txt", "r") 
  print("File opened")
  lines_list = fh.readlines()
except FileNotFoundError:
  print("File does not exist")

#read the whole text using readlines( ) method to get a list of lines

count = 1

fh2 = open("output.txt", "w")

#read each line and split the string to get date and the gas price
for i in lines_list:
    line_part1 = i.split('-')
    month = line_part1[0]
    day = line_part1[1]
    line_part2 = line_part1[2]
    line_part3 = line_part2.split(':')
    
    print('[',count,']')
    print("Month :",month)
    print("Day :",day)
    print("Year :", line_part3[0])
    print("Gas Price :",line_part3[1])
    
    #preparing the string to write to the file
    L = ["[",str(count),"]\n", "Month :",month,"\n","Day :",day,"\n","Year :", line_part3[0],"\n","Gas Price :",line_part3[1],"\n"]    

    #writing to the output.txt file
    fh2.writelines(L)

    count = count+1

fh2.close()
    




```
